## Loader

```python
transform = transforms.Compose([transforms.Resize([128,128]),
                                transforms.ToTensor(),
                           #     transforms.Normalize((0.5, ), (0.5, ))
])

data_train = datasets.MNIST(root = "../Dataset/MNIST/",
                            transform=transform,
                            train = True,
                            download = True)

data_test = datasets.MNIST(root="../Dataset/MNIST/",
                           transform = transform,
                           train = False)

train_loader = torch.utils.data.DataLoader(dataset=data_train,
                                           batch_size = 64,
                                           shuffle = True,
                                          drop_last=True)

valid_loader = train_loader

test_loader = torch.utils.data.DataLoader(dataset=data_test,
                                          batch_size = 64,
                                          drop_last=True,
                                          shuffle = True)
```

```python
transform = transforms.Compose([transforms.Resize([128,128]),
                                transforms.ToTensor(),
                      #          transforms.Normalize((0.5, ), (0.5, ))
])
datadir = '../Dataset/MNIST/myMNIST/'
# Specify x and y
mydata = TP.MyDataset(datadir + 'train_images_30c2000s.npy',
                      datadir + 'train_images_30c2000s.npy',
                      (28, 28),
                      transform,
                     img='y')
print(mydata.data['x'].shape, mydata.data['y'].shape)

# Specify sizes of train, valid and test sets
train_size = int(len(mydata) * 0.7)
valid_size = int(len(mydata) * 0.2)
test_size = int(len(mydata)) - train_size - valid_size
print(train_size, valid_size, test_size)

train_loader, valid_loader, test_loader = TP.split_loader(mydata, train_size, valid_size, test_size, 64)
```

```python
# Hyperparameters
t_args = TP.MyArgs(cuda=0, epochs=10, learning_rate=1e-4, criterion=nn.MSELoss(reduction='sum'))
s_args = TP.MyArgs(cuda=0, epochs=100, learning_rate=1e-5, criterion=nn.MSELoss())
```

```python
# Model and Loss
imgencoder = TP.ImageEncoder(batchnorm=False, latent_dim=8).to(t_args.device)
imgdecoder = TP.ImageDecoder(latent_dim=8, active_func=nn.LeakyReLU(inplace=True)).to(t_args.device)
csiencoder = TP.CsiEncoder(batchnorm=False, latent_dim=8).to(t_args.device)
trainer = TP.TrainerTS(imgencoder, imgdecoder, csiencoder,
                                   t_args, s_args, 
                                   train_loader, valid_loader, test_loader,
                                   temperature=1)
```

```python
# Load from checkpoints

imgencoder.load_state_dict(torch.load('../Models/AE-MNIST-ep530_En.pth'))
imgdecoder.load_state_dict(torch.load('../Models/AE-MNIST-ep530_De.pth'))
```

```python
# When changing learning rate
t_args = TP.MyArgs(cuda=0, epochs=50, learning_rate=1e-4, criterion=nn.MSELoss())
s_args = TP.MyArgs(cuda=0, epochs=100, learning_rate=1e-7, criterion=nn.MSELoss())
trainer.args={'t':t_args, 's':s_args}
```

```python
# Training and Saving
trainer.train_teacher()
```

```python
# Plot loss curves
%matplotlib inline
trainer.plot_teacher_loss()
```

```python
  #=====================test============================
trainer.test_teacher(mode='train')
```
